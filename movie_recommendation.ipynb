```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

 Data Collection and Pre-processing

```python
cols=['id', 'Name', 'Genre', 'Language', 'Summary', 'Popularity', 'Release date', 'Vote average', 'Rating']
movies_data=pd.read_csv("movies_data.csv",names=cols)
```

```python
movies_data=movies_data.drop(['id','Release date','Vote average'],axis=1)
```

```python
movies_data=movies_data.drop([0],axis=0)
```

```python
idx=0
movies_data.insert(idx, 'index', value=np.arange(len(movies_data)))
```

```python
movies_data
```

```python
features=['Name', 'Genre', 'Language', 'Summary', 'Popularity', 'Rating']
```

```python
#replacing null values with null strings
for i in features:
  movies_data[features]=movies_data[features].fillna('')
```

```python
#Combining selected features
combined_features=movies_data['Genre']+' '+movies_data['Language']+' '+movies_data['Summary']+' '+movies_data['Popularity']+' '+movies_data['Rating']
```

```python
print(combined_features)
```

```python
#Converting text to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
# Getting similarity Scores
similar=cosine_similarity(feature_vectors)
```

```python
print(similar)
```

```python
print(similar.shape)
```

```python
# Getting movie name from user
user_movie=input("Enter movie name: ")
```

```python
#List of movies
list_movies=movies_data['Name'].tolist()
print(list_movies)
```

```python
#finding closest matches of movie
close_movie_match=difflib.get_close_matches(user_movie, list_movies)
print(close_movie_match)
```

```python
close_match=close_movie_match[0]
print(close_match)
```

```python
index_movie = movies_data[movies_data.Name == close_match]['index'].values[0]
print(index_movie)
```

```python
index_sort=sorted(list(enumerate(similar[index_movie])), reverse=True, key= lambda x:x[1])
print(index_sort)
```

```python
for i in index_sort[1:30]:
  print(movies_data.iloc[i[0]].Name)
```

```python
user_movie=input("Enter movie name: ")

list_movies=movies_data['Name'].tolist()

close_movie_match=difflib.get_close_matches(user_movie, list_movies)

close_match=close_movie_match[0]

index_movie = movies_data[movies_data.Name == close_match]['index'].values[0]

index_sort=sorted(list(enumerate(similar[index_movie])), reverse=True, key= lambda x:x[1])

print("Movies recommended: \n")

for i in index_sort[0:30]:
  print(movies_data.iloc[i[0]].Name)
```
